```python
## Binary Classifier for digit 4
y_train_4 = (y_train==4)
y_test_4 = (y_test==4)
```

```python
from sklearn.linear_model import SGDClassifier
classifier = SGDClassifier(random_state=42)
classifier.fit(X_train, y_train_4)

```

```python
output_01= classifier.predict(X_test) 
```

```python
output_01
```

```python
np.unique(output_01)
```

```python
import sklearn.model_selection
sklearn.model_selection.cross_val_score(classifier,X_train,y_train_4,cv=3,scoring="accuracy")
```

```python
from sklearn.model_selection import cross_val_predict
y_train_pred = cross_val_predict(classifier,X_train,y_train_4,cv=3)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_train_4,y_train_pred) 
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
precision_score(y_train_4, y_train_pred)
```

```python
recall_score(y_train_4, y_train_pred)
```

```python
f1_score(y_train_4, y_train_pred)
```
